```python
import pandas as pd
import statistics
import plotly.express as px
import plotly.graph_objects as go
import numpy
import plotly.figure_factory as ff
import random
import csv
```

```python
df = pd.read_csv("IMDB Movies.csv")
imdb_title_id = df["imdb_title_id"].tolist()
avg_vote = df["avg_vote"].tolist()
votes = df["votes"].tolist()
reviews_from_users = df["reviews_from_users"].tolist()
year = df["year"].tolist()
country = df["country"].tolist()
budget = df["budget"].tolist()
```

Graph Based On Movie Corresponding To It's Year.
The Later The Movie Came Or The Bigger The Movie ID Is, The Newer The Movie Is.

```python
fig = px.scatter(x=imdb_title_id, y=year, color=avg_vote, title="Data Frame(Year-Movie)")
fig.layout.update({
    "yaxis": {"title": "Release Year Of Movie"},
    "xaxis": {"title": "Movie ID"}
})
fig.show()
```

```python
# Finding Mean, Median, Mode ANd Standard Deviation For The Movie Released Year
print(f"Mean For The Year Is: {statistics.mean(year)}")
print(f"Mode For The Year Is: {statistics.mode(year)}")
print(f"Median For The Year Is: {statistics.median(year)}")
print(f"Standard Deviation For The Year Is: {statistics.stdev(year)}")
```

As per the results, most movies were made in 2014, and 2010 is the middle set of all the data.

```python
fig = ff.create_distplot([year], ["Year"], colors=["royalblue"], show_hist=False)

# Updating Graph
fig.layout.update({
                    "title": "Release Year Of Movies",
                    "xaxis": {"title": "Year"},
                    "yaxis": {"title": "Movies Released"},
                 })

year_mean = statistics.mean(year)
fig.add_traces(go.Scatter(x=[year_mean, year_mean], y=[0, 0.06], mode="lines", name="Mean"))

# Plotting Graph
fig.show()
```

The Graph Is Not Following Normal Distrinution.
The Movies Are Released According To Year Or Any Parameter.

```python
# Finding Mean, Median, Mode For Average Vote Of The Movie
avg_vote_mean = statistics.mean(avg_vote)
avg_vote_mode = statistics.mode(avg_vote)
avg_vote_median = statistics.median(avg_vote)
avg_vote_std_deviation = statistics.stdev(avg_vote)

print("Mean For Average Rating Is: ", avg_vote_mean)
print("Mode For Average Rating Is: ", avg_vote_mode)
print("Median For Average Rating Is: ", avg_vote_median)
print("Standard Deviation For Average Rating Is: ", avg_vote_std_deviation)
```

```python
# Finding Mean, Median, Mode For Voting Of The Movie
vote_mean = statistics.mean(votes)
vote_mode = statistics.mode(votes)
vote_median = statistics.median(votes)
vote_std_deviation = statistics.stdev(votes)

print("Mean For Voting Of Movie Is: ", vote_mean)
print("Mode For Voting Of Movie Is: ", vote_mode)
print("Median For Voting Of Movie Is: ", vote_median)
print("Standard Deviation For Voting Of Movie Is: ", vote_std_deviation)
```

```python
fig = ff.create_distplot([avg_vote], ["Average Voting"], show_hist=False)
fig.add_traces(go.Scatter(x=[avg_vote_mean, avg_vote_mean], y=[0, 0.4], mode="lines", name="Mean"))
fig.add_traces(go.Scatter(x=[avg_vote_std_deviation, avg_vote_std_deviation], y=[0, 0.4], mode="lines", name="Standard Deviation"))
fig.layout.update({
    "title": "Average Voting Of Movies",
    "xaxis": {"title": "Average Voting"},
    "title_font_color": "royalblue"
})
fig.show()
```

The graph is not normally distributed. People have given rating near to 6 and after rating of 4 the graph starts following a normal distribution which means average lowest rating is 3-4 and average highest rating is 6-8.
Rare people have given rating adter 8 as the graph starts coming down.

Now, we will draw the graph corresponding to votes given to it and release year of movie.

```python
fig = px.scatter(x=year, y=votes, color=avg_vote)
fig.layout.update({
    "title": "Votes Given To Movies In Years",
    "xaxis": {"title": "Years"},
    "yaxis": {"title": "Votes"},
})
fig.show()
```

The data is scattered here and there and has a significant difference.

```python
# Checking Its Normal Distribution
fig = ff.create_distplot([votes], ["Votes"], show_hist=False)
fig.layout.update({
    "title": "Votes Given By All People",
    "xaxis": {"title": "Votes"}
})
fig.show()
```

```python
# Finding The IQR
q1 = df["votes"].quantile(0.25)
q3 = df["votes"].quantile(0.75)
iqr = q3 - q1

print("Quantile 1 And 3 is: "+ str(q1) + " and " + str(q3))
print("IQR is: ", iqr)
```

```python
# Finding Lower And The Upper Whiskar
lower_whiskar = q1 - (1.5 * iqr)
upper_whiskar = q3 + (1.5 * iqr)

print("Lower Whiskar Is: ", lower_whiskar)
print("Upper Whiskar Is: ", upper_whiskar)
```

```python
# Creating New Df
new_df = df[df["votes"] < upper_whiskar]
voting_list = new_df["votes"].tolist()

print(f"Mean For The New Voting List Is: {statistics.mean(voting_list)}")
print(f"Mode For The New Voting List Is: {statistics.mode(voting_list)}")
print(f"Median For The New Voting List Is: {statistics.median(voting_list)}")
print(f"Standard Deviation For The New Voting List Is: {statistics.stdev(voting_list)}")
```

```python
fig = ff.create_distplot([voting_list], ["Votes"], show_hist=False)
fig.layout.update({
    "title": "Votes Given By All People",
    "xaxis": {"title": "Votes"}
})
fig.show()
```

```python
mean_list = []
for i in range(0, 1000):
  sample_list = []
  for j in range(0, 100):
    sample_list.append(random.choice(voting_list))
  mean_list.append(statistics.mean(sample_list))
```

```python
final_mean = statistics.mean(mean_list)
final_std_deviation = statistics.stdev(mean_list)

print("New Mean Is: ", final_mean)
print("New Standard Deviation Is: ", final_std_deviation)
```

```python
fig = ff.create_distplot([mean_list], ["Sample Of Data"], show_hist=False)
fig.add_traces(go.Scatter(x=[final_mean, final_mean], y=[0, 0.00008], mode="lines", name="Mean"))
fig.layout.update({
    "title": "Sample Data (100 Data Set)",
    "xaxis": {"title": "Votes"}
})
fig.show()
```

Till here, we can conclude that:-
    
    Most movies were made in 2014.
    The greater the movie ID, the newer the movie launched.
    The movies are not evenly created and are not following normal distribution.
    The average vote given by most of the people is 6.4.
    The votes given to the movie is greater and due to which there mean is also increased because fo the outliers.
    After Removing th eoutliers and reshaping the data, the mean and the standard deviation decreased.
    And at last, by taking the samples the votes are undergoing normal distribution.
